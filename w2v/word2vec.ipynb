```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec.load('/gdrive/MyDrive/colab/w2v/ko/ko.bin')
```

```python
print('1. 한국 - 서울 + 파리')
print(model.wv.similar_by_vector(model.wv['한국'] - model.wv['서울'] + model.wv['파리'], topn=3))
print()
print('2. 컴퓨터 - 과학 + 인문학')
print(model.wv.similar_by_vector(model.wv['컴퓨터'] - model.wv['과학'] + model.wv['인문학'], topn=3))
print()
print('3. 사랑 - 이별')
print(model.wv.similar_by_vector(model.wv['사랑'] + model.wv['이별'], topn=3))
print()
print('4. 충청북도 - 청주 + 강원도')
print(model.wv.similar_by_vector(model.wv['충청북도'] - model.wv['청주'] + model.wv['강원도'], topn=3))
print()
print('5. 인간 - 사랑')
print(model.wv.similar_by_vector(model.wv['인간'] - model.wv['사랑'], topn=3))
print()
print('6. 미술 - 물감 + 음악')
print(model.wv.similar_by_vector(model.wv['미술'] - model.wv['물감'] + model.wv['음악'], topn=3))
```
